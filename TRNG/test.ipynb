```python
import time
from pynput import mouse

x_coo = []
y_coo = []

def collect_mouse_entropy_simple():
    controller = mouse.Controller()
    x, y = controller.position
    x_coo.append(x)
    y_coo.append(y)
    time.sleep(0.01)
```

```python
del x_coo[:]
del y_coo[:]
print("Inizia a mouovere il mouse")
time.sleep(2)
t_end = time.time() + 15
while time.time() < t_end:
    collect_mouse_entropy_simple()
```

```python
import numpy as np

x_m = np.asarray(x_coo, dtype=np.float32)
y_m = np.asarray(y_coo, dtype=np.float32)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_m, y_m, s=1)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Mouse positions")
plt.show()
```

```python
import numpy as np

bit_m = np.zeros(shape=(len(x_m)))
for i in range(len(x_m)):
    temp = (x_m[i] + y_m[i]).astype(int)
    bit_m[i] = 0 if temp % 2 == 0 else 1
```

```python
ones = 0
for i in range(len(bit_m)):
    if bit_m[i] == 1:
        ones += 1

print(ones)
```

```python
zeros = len(bit_m) - ones
```

```python
print(zeros, ones)
print(zeros - ones)
```

```python
plt.figure(figsize=(20, 4))  # larghezza=12, altezza=4 pollici
plt.plot(range(len(bit_m)), bit_m)
plt.ylim(-0.2, 1.2)
plt.xlabel("Indice")
plt.ylabel("Valore")
plt.title("Bit nel tempo")
plt.grid(True)
plt.show()
```
